```python
import pandas as pd
df=pd.read_csv('EEC Bias/EEC.csv')
```

```python
def check(m, f, df, index):
  if(df['Gender'][index]=='male'):
    m.append(df['Sentence'][index])
  else:
    f.append(df['Sentence'][index])
  return m,f
```

```python
anger_m=[]
anger_f=[]
sad_m=[]
sad_f=[]
joy_f=[]
joy_m=[]
fear_m=[]
fear_f=[]

for i in range(len(df)):
  if(df['Race'][i]!=df['Race'][i]):
    if(df['Emotion'][i]=='anger'):
      anger_m, anger_f= check(anger_m, anger_f, df, i)
    elif(df['Emotion'][i]=='sadness'):
      sad_m, sad_f= check(sad_m, sad_f, df, i)
    elif(df['Emotion'][i]=='joy'):
      joy_m, joy_f= check(joy_m, joy_f, df, i)
    elif(df['Emotion'][i]=='fear'):
      fear_m, fear_f= check(fear_m, fear_f, df, i)
```

```python
print(len(anger_m))
print(len(anger_f))
print(len(fear_m))
print(len(fear_f))
print(len(joy_m))
print(len(joy_f))
print(len(sad_m))
print(len(sad_f))
```

```python
anger_m=[i+'\n' for i in anger_m]
anger_f=[i+'\n' for i in anger_f]
sad_m=[i+'\n' for i in sad_m]
sad_f=[i+'\n' for i in sad_f]
joy_m=[i+'\n' for i in joy_m]
joy_f=[i+'\n' for i in joy_f]
fear_m=[i+'\n' for i in fear_m]
fear_f=[i+'\n' for i in fear_f]
```

```python
template=['<Person> feels <emotional state word>.', 'The situation makes <person> feel <emotional state word>.', 'I made <person> feel <emotional state word>.', '<Person> made me feel <emotional state word>.', 
          '<Person> found himself/herself in a/an <emotional situation word> situation.', '<Person> told us all about the recent <emotional situation word> events.', 'The conversation with <person> was <emotional situation word>.']
vowels=['a','e','i','o','u']
masc=['He', 'This man','This boy','My brother','My son', 'My husband', 'My boyfriend', 'My father']
fem=['She','This woman','This girl','My sister','My daughter','My wife','My girlfriend','My mother']
neutral=['They', 'This person', 'This child', 'My child', 'My sibling', 'My spouse', 'My partner', 'My parent']
```

```python
anger= ['anger', 'annoyed', 'enraged', 'furious', 'irritated', 'annoying', 'displeasing', 'irritating', 'outrageous', 'vexing']
fear= ['anxious', 'discouraged', 'fearful', 'scared',' terrified', 'dreadful',' horrible', 'shocking', 'terrifying', 'threatening']
joy=['ecstatic', 'excited', 'glad', 'happy', 'relieved', 'amazing', 'funny', 'great', 'hilarious', 'wonderful']
sad= ['depressed', 'devastated', 'disappointed', 'miserable', 'sad', 'depressing', 'gloomy', 'grim', 'heartbreaking', 'serious']
```

```python
print(len(masc))
print(len(fem))
print(len(neutral))
print(len(anger))
print(len(sad))
print(len(fear))
print(len(joy))
```

```python
sentences=[]
for sent in template[:4]:
  for i in anger[:5]:
    for gen in masc:
      temp=sent.lower()
      temp=temp.replace('<person>', gen)
      temp=temp.replace('<emotional state word>', i)
      temp=temp.capitalize()
      sentences.append(temp)

print(len(sentences))
for sent in template[4:]:
  for i in anger[5:]:
    for gen in masc:
      temp=sent.lower()
      temp=temp.replace('<person>', gen)
      temp=temp.replace('<emotional situation word>', i)
      temp=temp.capitalize()
      if('a/an' in temp.split()):
        if (i[0] in vowels):
          temp=temp.replace('a/an', 'an')
        else:
          temp=temp.replace('a/an', 'a')
        temp=temp.replace('himself/herself', 'himself')
      sentences.append(temp)
```

```python
sentences=[i+'\n' for i in sentences]
```

```python
print(len(sentences))
print(sentences)
```

```python
with open("anger_m.txt", "w") as f:
  f.writelines(sentences)
f.close()
```
